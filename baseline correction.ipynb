```python
from biosppy.signals import ecg
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
ecg_data_file = 'D:/data/iBeat/eplab/Subject49_data.hdf'
ecg_data = pd.read_hdf(ecg_data_file,'eplab')
ecg_data = ecg_data[ (ecg_data.utc_adjusted > 1528139300) & (ecg_data.utc_adjusted < 1528139315)]
```

```python
SR = 976.942

out = ecg.ecg(signal=ecg_data.II, sampling_rate=SR,show = False)
ecg_sig_II = out['filtered']
```

```python
plt.figure(figsize=(10,5))
plt.plot(ecg_data.utc_adjusted, ecg_data.II,label = 'Raw data')
plt.plot(ecg_data.utc_adjusted, ecg_sig_II,label = 'Baseline correction')
plt.ylabel('Amplitude/mV')
plt.xlabel('Time')
plt.legend(prop={'size': 10})
```
